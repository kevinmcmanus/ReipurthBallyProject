```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 50, 2000)
```

```python
omega = 2.0 # radians per second
```

```python
fig = plt.figure(figsize=(8,8))
ax1 = plt.subplot(111)
ax1.plot(t, np.cos(omega*t))
ax1.plot(t, np.sin(omega*t))
```

```python
a=8.2; b=2.1
fig = plt.figure(figsize=(8,8))
ax1 = plt.subplot(111)
ax1.plot( a*np.cos(omega*t), b*np.sin(omega*t))
ax1.axis('equal')
```

```python
res = (a*np.cos(omega*t))**2 + (b*np.sin(omega*t))**2
```

```python
fig = plt.figure(figsize=(8,8))
ax1 = plt.subplot(111)
ax1.plot( t, res)
```

```python
res2 = (np.cos(omega*t)/a)**2 + (np.sin(omega*t)/b)**2
fig = plt.figure(figsize=(8,8))
ax1 = plt.subplot(111)
ax1.plot( t, res2)
```
