```python
import numpy as np
def  image_stats(img_data):
    s = img_data.shape
    stats = {'nrows':s[0], 'ncols': s[1],
        'min':np.nanmin(img_data), 'max': np.nanmax(img_data),
             'mean': round(np.nanmean(img_data),1), 'stddev':round(np.nanstd(img_data),1),'med':np.nanmedian(img_data),
             'dtype':img_data.dtype, 'n_nans':np.isnan(img_data).sum(),
             'n_32768': (img_data == -32768).sum()}
    
    return stats
```

```python
import matplotlib.table as mplTable
import matplotlib.gridspec as gridspec
from astropy import visualization as aviz

def image_summary(img_data, title=None, cmap='gray', figsize=(9,9)):
    fig = plt.figure(figsize=figsize)
    gs4 = gridspec.GridSpec(ncols=2, nrows=2, figure=fig)
    ax1 = fig.add_subplot(gs4[:,0])
    stats = image_stats(img_data)
    #row_labels = list(stats.keys())
    cells = [[v, str(stats[v])] for v in stats]
    show_image(data, fig=fig, ax=ax1, cmap=cmap)

    ax_hist = fig.add_subplot(gs4[0,1])
    zz = aviz.hist(img_data.flatten(), bins=250, color=cmap, log=True)

    ax_tbl = fig.add_subplot(gs4[1,1])
    ax_tbl.axis('tight')
    ax_tbl.axis('off')
    ax_tbl.table(cellText=cells, loc='center',
                 colLabels=['Measure', 'Value'])
    if title is not None:
        fig.suptitle(title)
    
    plt.tight_layout()

    
```
