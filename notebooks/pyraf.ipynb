```python
fig, axs=plt.subplots(figsize=(16,8), ncols=2, nrows=1)
dim = ['x','y']
for d, ax in zip(dim, axs):
    ax.scatter(res_df[d+'_ref'],res_df[d+'_in'], label='Actual')
    ax.scatter(res_df[d+'_ref'],res_df[d+'_fit'], label='Fitted')
    ax.set_title(f'Dimension: {d}')
    ax.legend()
    ax.set_ylabel('Pixel')
    ax.set_xlabel('Reference Pixel')
```
